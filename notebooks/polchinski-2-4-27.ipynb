```python
from sympy import *
init_printing(use_unicode=True)
```

```python
z = symbols('z')
f, g, t = symbols('f, g, t', cls=Function)
g = g(z)
f = f(g)
```

```python
def schwarzian(f, z):
    return ( \
        2 * diff(f, z, 3) * diff(f, z) - \
        3 * diff(f, z, 2) ** 2) / (2 * diff(f, z) ** 2)
```

```python
lhs = schwarzian(g, z) + diff(g, z) ** 2 * schwarzian(f, g)
lhs
```

```python
rhs = schwarzian(f, z)
rhs
```

```python
Eq(lhs, rhs).simplify()
```
